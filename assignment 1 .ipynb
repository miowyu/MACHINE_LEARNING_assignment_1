```python
def update_rule_LM(data,label,max_step=10,c_step=1):
    an = data.shape[1]
    n = data.shape[0]
    #所有參數估計值預設為0
    beta = np.array([0]*an,dtype="float")
    #常數c預設值也是0
    c = 0
    #suffle
    r_list = list(range(n))
    random.shuffle(r_list)
    data = data.iloc[r_list,:]
    label = label.iloc[r_list]
    #開始建立模型:
    for i in range(n):
        value = sum(data.iloc[i,:]*beta)+c
        if(value==0):
            value = -1
        #一個sample只跌代max_step次，超過還是錯的就直接跳下一個sample
        k = 0
        while ((value*label.iloc[i]<0)&(k<max_step)):
            #所有beta都要檢查更新方向
            for z in range(an):
                #計算這個bate是要增加還是減少
                if data.iloc[i,z] > 0:
                    beta[z] += label.iloc[i]
                else:
                    beta[z] -= label.iloc[i]
            c += label.iloc[i]*c_step
            k+=1
            value = sum(data.iloc[i,:]*beta)+c
            
            
    predict = []
    for i in range(n):
        value = sum(data.iloc[i,:]*beta)+c
        if(value>0):
            predict.append(1)
        else:
            predict.append(-1)
            
    predict=pd.DataFrame({"predict":predict})
    predict.index = r_list
    predict = predict.sort_index()["predict"]

    return(predict,c,beta)
```

```python
def least_squre(data,label):
    b0 = np.ones((data.shape[0], 1))
    newdata = np.hstack((data, b0))
    est_beta = np.linalg.inv(newdata.T @ newdata) @ newdata.T @ label
    predict = newdata @ est_beta
    predict_dict = {True:1,
                False:-1}
    predict = (pd.Series(predict)>0).map(predict_dict)
    return(est_beta,predict)
```

```python
#model_n建議奇數，偶數雖然也可以，但當voted票數相等時，會直接當成0
def perception_update_LM(data,label,model_n,method,max_step=10,c_step=1):
    #指定一個seed的列表，範圍從0~1000000
    seed_list = random.sample(range(0, 1000000), model_n)
    if method == "voted":
        predict_table = pd.DataFrame(columns = list(range(model_n)), index=list(range(data.shape[0])))
        for i in range(model_n):
            random.seed(seed_list[i])
            predict,c,beta= update_rule_LM(data,label,max_step=max_step,c_step=c_step)
            predict_table.iloc[:,i] = predict
        predict = []
        for i in range(data.shape[0]):
            if(sum(predict_table.iloc[i,:])>0):
                predict.append(1)
            else:
                predict.append(-1)
        predict = pd.Series(predict)
        return(predict_table,predict)
    elif method == "average":
        est_beta = np.zeros((1, data.shape[1]+1))
        for i in range(model_n):
            random.seed(seed_list[i])
            predict,c,beta= update_rule_LM(data,label,max_step=max_step,c_step=c_step)
            est_beta[0,0] = est_beta[0,0]+c/model_n
            est_beta[0,1:] = est_beta[0,1:]+beta/model_n
        
        for i in range(data.shape[0]):
            b0 = np.ones((data.shape[0], 1))
            newdata = np.hstack((b0,data))
            predict = est_beta @ newdata.T
            predict_dict = {True:1,
                        False:-1}
            predict = (pd.Series(predict[0])>0).map(predict_dict)
        return(est_beta,predict)
            
            
        
        
    else:
        print("method must be voted or average")
        return(0)
    
```

```python
def w_estimate(beta,data,slack):
    #算距離，因為beta包含常數項，所以分母要扣掉
    dist = np.absolute(data @ beta)/(((beta@beta)-beta[len(beta)-1]**2)**(1/2))
    judge = (dist > slack).astype(int)
    return(dist @ judge)
def update_rule_LM_rewrite(data,label,max_iter=5,c_step=1,slack=0):
    an = data.shape[1]
    n = data.shape[0]
    #所有參數估計值預設為6
    beta = np.array([6]*(an+1),dtype="float")
    #常數c預設值設為-an
    beta[-1] = -an
    
    #suffle
    r_list = list(range(n))
    random.shuffle(r_list)
    data = data.iloc[r_list,:]
    label = label.iloc[r_list]
    b0 = np.ones((data.shape[0], 1))
    newdata = np.hstack((data, b0))
    #計算起始線的W
    dist = w_estimate(beta,newdata,slack)
    #開始建立模型:
    iterate = 0
    keep = True
    accept = 0
    drop = 0
    while (iterate < max_iter)&keep:
        keep = False
        iterate +=1
        for i in range(n):
            value = sum(newdata[i,:]*beta)
            if(value==0):
                value = -1

            if(value*label.iloc[i]<0):
                #所有beta都要檢查更新方向
                
                for z in range(an):
                    #計算這個bate是要增加還是減少
                    newbeta = beta.copy()
                    if newdata[i,z] > 0:
                        newbeta[z] += label.iloc[i]
                        
                    else:
                        newbeta[z] -= label.iloc[i]

                    newdist = w_estimate(newbeta,newdata,slack)
                #看是否要進行疊代，每個參數更改後都進行一次
                if newdist < dist:
                    dist = newdist
                    beta = newbeta
                #看常數是否要進行疊代
                newbeta = beta.copy()
                newbeta[-1] += label.iloc[i]*c_step
                
                newdist = w_estimate(newbeta,newdata,slack)
                if newdist < dist:
                    dist = newdist
                    beta = newbeta
                    keep = True
                    accept+=1
                else:
                    drop+=1
                    
    print("stop at iterate:",iterate)
    predict = []
    for i in range(n):
        value = sum(newdata[i,:]*beta)
        if(value>0):
            predict.append(1)
        else:
            predict.append(-1)
            
    predict=pd.DataFrame({"predict":predict})
    predict.index = r_list
    predict = predict.sort_index()["predict"]

    return(predict,c,beta,drop,accept)
```

```python
#data+ minimizing the ||w||^2
random.seed(6111016)
predict_rewrite,c,beta,drop,accept = update_rule_LM_rewrite(data_x_normal,data_y,max_iter=20,c_step=2,slack=0)
rewrite_table = confusion_matrix(data_y, predict_rewrite)
print(sum(predict_rewrite))
print(rewrite_table)
print("accuracy is %.4f"%((rewrite_table[0,0]+rewrite_table[1,1])/sum(sum(rewrite_table))))
print(drop,accept)
```
